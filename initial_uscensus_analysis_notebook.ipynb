```python
import pandas as pd
import requests
```

```python
census_mode_share_url = "https://api.census.gov/data/2023/acs/acs5?get=group(B08006)&ucgid=pseudo(0400000US36$1400000)"

api_response = requests.get(census_mode_share_url)

b08006_col_names = list(
    pd.read_html(
        'https://api.census.gov/data/2023/acs/acs1/groups/B08006.html'
    )[0]['Label']
)

if api_response.status_code == 200:
    # Parse the JSON response
    data = api_response.json()

    # The rest of the rows contain the data
    rows = data[1:]
    
    # Create a DataFrame
    df = pd.DataFrame(rows, columns=b08006_col_names)
else:
    raise Exception(f"Failed to fetch data: {api_response.status_code}")

df.head()
```
